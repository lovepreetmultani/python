```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
tips.describe()
```

```python
sns.displot(tips['total_bill'],kde=False,bins=10) 
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.barplot(x='sex', y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True, split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips,color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

```python
flights.head() 
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc, annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
import seaborn as sns1
iris=sns1.load_dataset('iris')
iris.head()
```

```python
sns1.pairplot(iris)
```

```python
g=sns1.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
